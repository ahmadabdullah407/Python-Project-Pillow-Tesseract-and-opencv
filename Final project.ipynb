# Solution:

## Using small_img.zip file and searching for 'Christopher':

```python
import zipfile
from zipfile import ZipFile
from PIL import Image
import pytesseract
import cv2 as cv
import numpy as np

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
def open_zip(path):
    myzip = ZipFile(str(path))
    objlist = myzip.infolist()
    biglist=[]
    for obj in objlist:
        imgdict={}
        file=myzip.open(obj)
        image=Image.open(file)
        imgdict[obj.filename]=image
        text = pytesseract.image_to_string(image)
        imgdict['text']=text
        sublist=[imgdict,obj.filename]
        biglist.append[sublist]
    return biglist
def crop_faces(faces,pil_img):   
    flist=[]
    for x,y,w,h in faces:
        cropped_face=pil_img.crop((x,y,x+w,y+h))
        flist.append(cropped_face)
    return flist
def detect_faces(pil_img):
    cv_img = np.array(pil_img)
    faces = face_cascade.detectMultiScale(cv_img,1.35,4).tolist()
    return faces
def create_cs(faces):
    width_cs = 550
    height_cs = 110*int((np.ceil(len(faces)/5)))
    contact_sheet=Image.new('RGB',(width_cs,height_cs))
    x=0
    y=0
    for face in faces:
        face.thumbnail((110,110))
        contact_sheet.paste(face, (x, y) )
        if x+110 == contact_sheet.width:
            x=0
            y=y+110
        else:
            x=x+110
    return contact_sheet
def search(string,imgdict,filename):
    if string in imgdict['text']:
        print('Results found in file', filename)
        try:
            faces = detect_faces(imgdict[filename])
            imgdict['faces'] = crop_faces(faces,imgdict[filename])
            face_cs=create_cs(imgdict['faces'])
            display(face_cs)
        except:
            print('But there were no faces in that file!')
    else:
        pass
in_string=input("Enter text to search :")
imglist=open_zip("readonly/small_img.zip")
for sublist in imglist:
    search(in_string,sublist[0],sublist[1])
```

## Using images.zip file and searching for 'Mark':

```python
in_string=input("Enter text to search :")
imglist=open_zip("readonly/images.zip")
for sublist in imglist:
    search(in_string,sublist[0],sublist[1])
```
